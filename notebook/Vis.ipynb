```python
import matplotlib.pyplot as plt
import numpy as np
import math
import csv
%matplotlib inline
```

```python
def rosenbrock2(x, y):
    value = (1 - x) ** 2 + 100 * (y - x ** 2) ** 2
    return value;
```

```python
x = np.arange(-5.0,5.0,0.1)
y = np.arange(-5.0,10.0,0.1)
X,Y = np.meshgrid(x, y)
Z = rosenbrock2(X, Y)

with open('path.csv', newline='') as csvfile:
    raw_csv = csv.reader(csvfile, delimiter='|')
    path = [[float(i.replace(',', '.')) for i in row] for row in raw_csv]
pathx, pathy = zip(*path)

plt.contourf(X, Y, Z, 250, cmap=plt.cm.coolwarm)
plt.plot(pathx, pathy, 'k', pathx, pathy, 'ro')
plt.show()
```
